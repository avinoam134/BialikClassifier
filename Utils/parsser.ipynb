```python
from trankit import Pipeline

p = Pipeline("hebrew")
```

```python
import json
POEMS_PATH = "/../Poems_And_Outputs/Poems_Unparsed.JSON"

def parse_poems(file_path, encoding='utf-8'):
    with open(file_path, "r", encoding=encoding) as file:
        data = json.load(file)
    return data

data = parse_poems(POEMS_PATH)
```

```python
def main():
    res = []
    for i in range(150,172):
      temp = {}
      temp['name'] = data[i]['poem']
      temp['year'] = data[i]['year']
      temp['published'] = data[i]['published']
      temp['content'] = p(data[i]['content'])
      res.append(temp)
      print("done: " + str(i))
    resJsonData = json.dumps(res,ensure_ascii=False ,indent=4)
    with open('output150-171.json', 'w' ,encoding="utf-8") as json_file:
      json_file.write(resJsonData)
      print(json_file)
```

```python
main()
```

```python
import os

# List to store combined data
combined_data = []
generic_file_path = './../Poems_And_Outputs/output'
# File paths of the JSON files
file_paths = [f"{generic_file_path}0-29.json", f"{generic_file_path}30-59.json", f"{generic_file_path}60-89.json", f"{generic_file_path}90-119.json", f"{generic_file_path}120-149.json", f"{generic_file_path}150-171.json"]

# Read data from each file and append to combined_data list
for file_path in file_paths:
    with open(file_path, 'r') as file:
        data = json.load(file)
        combined_data.extend(data)

# Write combined data to a new JSON file
output_file_path = "./../Poems_And_Outputs/allParsd.json"
with open(output_file_path, 'w',encoding="utf-8") as output_file:
    json.dump(combined_data, output_file, ensure_ascii=False , indent=4)

print("Combined data written to:", output_file_path)
```
